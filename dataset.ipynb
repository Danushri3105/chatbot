```python
import csv
import random

def create_and_store_dataset(csv_file_name, num_samples=100):
    header = ['ID', 'Feature1', 'Feature2', 'label']

    data = [[i, random.uniform(0.10, 10.0), random.uniform(0, 10), random.choice([0, 1])] for i in range(1, num_samples + 1)]

    with open(csv_file_name, 'w', newline='') as csv_file:
        csv_writer = csv.writer(csv_file)
        csv_writer.writerow(header)
        csv_writer.writerows(data)
        print(f"Dataset with {num_samples} samples created and stored in '{csv_file_name}'.")

def split_and_store_data(csv_file_name, train_file_name, test_file_name, test_ratio=0.2):
    with open(csv_file_name, 'r') as csv_file:
        csv_reader = csv.reader(csv_file)
        header = next(csv_reader)
        data = list(csv_reader)
        num_samples = len(data)
        num_test_samples = int(num_samples * test_ratio)
        num_train_samples = num_samples - num_test_samples
        train_data = data[:num_train_samples]
        test_data = data[num_train_samples:]

        with open(train_file_name, 'w', newline='') as train_file:
            csv_writer = csv.writer(train_file)
            csv_writer.writerow(header)
            csv_writer.writerows(train_data)

        with open(test_file_name, 'w', newline='') as test_file:
            csv_writer = csv.writer(test_file)
            csv_writer.writerow(header)
            csv_writer.writerows(test_data)

# Example usage:
create_and_store_dataset('dataset.csv')
split_and_store_data('dataset.csv', 'train_data.csv', 'test_data.csv')
```

```python
import pandas as pd

file_path='/content/Sample - Superstore.csv'
df=pd.read_csv(file_path,encoding='latin-1')

print("First few rows of the dataframe:")
print(df.head())
print()
df['year']=pd.to_datetime(df['Order Date'],dayfirst=True).dt.year

#filter
filtered_data=df[(df['year']==2017) & (df['City']== 'Los Angeles')]
#calculate
mean_sales=filtered_data['Sales'].mean()
mediam_sales=filtered_data['Sales'].median()
std_dev_sales=filtered_data['Sales'].std()

#print
print(f"mean sales for 2017 in ohio city:{mean_sales:2f}")
print(f"mediam sales for 2017 in ohio city:{mediam_sales}")
print(f"standard deviation of sales for 2017 in ohin city:{std_dev_sales}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#sample

data={
    'month': pd.date_range (start='2023-01-01', periods=12, freq='M'),
 'sales': [200, 220, 210, 190, 180, 160, 150, 140, 130, 120, 110, 100],
 'marketing_spend': [50,55,60,58,52,50,45,40,35,30,28,25],
 'customer_reviews': [4.5,4.6,4.6,4.4,4.2,4.0,3.8,3.7,3.6,3.5,3.4,3.3],
    'customer_sales': [180, 185, 190, 200, 210, 220, 230, 240, 250, 260,270,280]

}
#create dataframe
df=pd.DataFrame(data)
print(df)

```

```python
plt.figure(figsize=(10,6))
plt.subplot(3,2,1)
plt.plot(df['month'],df['sales'], marker='o')
plt.title('monthly sales')
plt.ylabel('sales')
```

```python
plt.subplot(3,1,2)
plt.plot(df['month'], df['marketing_spend'], marker='o',color='orange')
plt.title('monthly marketing spend')
plt.ylabel('marketing spend')
```

```python
plt.subplot(3,1,3)
plt.plot(df['month'], df['customer_reviews'], marker='o', color='green')
plt.title(' customer reviews')
plt.ylabel('customer reviews')

plt.tight_layout()
plt.show()

correlation=df.corr()
print(correlation)

```

```python
print("\nDiagnostic Analysis:")
print(f"sales are strongly correlated with marketing_spend: {correlation['sales']['marketing_spend']:.2f}")
print(f"sales are strongly correlated with customer reviews: {correlation['sales']['customer_reviews']:.2}f")
print(f"sales are strongly correlated with customer_sales: {correlation ['sales']['customer_sales']:.2f}")

if correlation['sales']['marketing_spend']>0.5:
  print("\nobservation: a decrease in marketing spend might be contributing to the sales drop.")

if correlation['sales']['customer_reviews']>0.5:
  print("\nobservation: a decrease in customer satisfaction (lower version) might be contributing to the sales drop.")

if correlation['sales']['customer_sales']>0.5:
  print("\nobservation: Increased customer_sales might be drawing be drawing customers away.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Titanic dataset from Seaborn's built-in datasets
df = sns.load_dataset('titanic')

# Display the first few rows of the dataset to understand its structure
print(df.head())

# Display basic information about the dataset, including column names and data types
print(df.info())

# Display summary statistics for all columns, including categorical and numeric
print(df.describe(include='all'))

# Bar plot of the number of passengers by class
plt.figure(figsize=(5, 5))  # Set the figure size
# Count the number of passengers in each class and create a bar plot
df['class'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Number of Passengers by Class')  # Add a title to the plot
plt.xlabel('Class')  # Label for the x-axis
plt.ylabel('Number of Passengers')  # Label for the y-axis
plt.show()  # Display the plot

# Pie chart of survival rates
plt.figure(figsize=(5, 5))  # Set the figure size
# Count the number of survivors and non-survivors and create a pie chart
survival_counts = df['survived'].value_counts()
plt.pie(survival_counts, labels=['Not Survived', 'Survived'], autopct='%1.1f%%', colors=['lightcoral', 'lightgreen'])
plt.title('Survival Rates')  # Add a title to the plot
plt.show()  # Display the plot

# Scatter plot of age vs fare, colored by survival status
plt.figure(figsize=(5, 6))  # Set the figure size
# Create a scatter plot with age on the x-axis, fare on the y-axis, and color indicating survival status
scatter = plt.scatter(df['age'], df['fare'], c=df['survived'], cmap='coolwarm', alpha=0.6)
plt.colorbar(scatter, label='Survived (0=No, 1=Yes)')  # Add a color bar to show the mapping of colors to survival status
plt.title('Age vs Fare by Survival Status')  # Add a title to the plot
plt.xlabel('Age')  # Label for the x-axis
plt.ylabel('Fare')  # Label for the y-axis
plt.show()  # Display the plot

# Boxplot of age distribution by class
plt.figure(figsize=(5, 6))  # Set the figure size
# Create a boxplot to show the distribution of age across different classes
sns.boxplot(x='class', y='age', data=df, palette='pastel')
plt.title('Age Distribution by Class')  # Add a title to the plot
plt.xlabel('Class')  # Label for the x-axis
plt.ylabel('Age')  # Label for the y-axis
plt.show()  # Display the plot

# Count plot of survival by sex
plt.figure(figsize=(5, 6))  # Set the figure size
# Create a count plot to show the number of survivors and non-survivors, broken down by sex
sns.countplot(x='survived', hue='sex', data=df, palette='viridis')
plt.title('Survival Count by Sex')  # Add a title to the plot
plt.xlabel('Survived')  # Label for the x-axis
plt.ylabel('Count')  # Label for the y-axis
plt.show()  # Display the plot
```
